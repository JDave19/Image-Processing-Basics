```python
value=50
hsv = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

lim = 255 - value
v[v > lim] = 255
v[v <= lim] += value

f_hsv = cv2.merge((h,s,v))
b_img = cv2.cvtColor(f_hsv, cv2.COLOR_HSV2BGR)
cv2.putText(b_img, text = 'Brightness = 50', org = (50, 500), fontFace = cv2.FONT_HERSHEY_DUPLEX,fontScale = 0.5, color = (225, 255, 255), thickness = 1)
dis = cv2.hconcat([b_img, img2])
cv2_imshow(dis)
```
