```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/lenna.png')
img = 0.289*image[:,:,0] + 0.587*image[:,:,1] + 0.114*image[:,:,2]

mean, var = 0,60
sigma = var ** 0.5
gaussian = np.random.normal(mean, sigma, (img.shape[0],img.shape[1])) 
# Add the Gaussian noise to the image
img_gauss = img + gaussian
```

```python
plt.figure(figsize = (20,20))
plt.subplot(121), plt.imshow(img, cmap = 'gray'), plt.title("Original image")
plt.subplot(122), plt.imshow(img_gauss, cmap = 'gray'), plt.title("Gaussian noisy image")
```
