```python
#Libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

```python
# original image
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/lenna.png')
img = 0.289*img[:,:,0] + 0.587*img[:,:,1] + 0.114*img[:,:,2]
cv2_imshow(img)
```

```python
fft =  np.fft.fftshift(img)
plt.imshow(np.log1p(np.abs(fft)),cmap='gray')
plt.axis('off')
plt.show()
```

```python
# image in frequency domain 
# the function fftshift shifts zero and low frequency components to the center.
# the function fft2 does 2D Fourier transform 
img_fft = np.fft.fftshift(np.fft.fft2(img))
plt.figure(figsize = (5,5))
plt.imshow(np.log1p(np.abs(img_fft)),cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Filter: Low pass filter
rows,cols = img.shape
H = np.zeros((rows,cols), dtype=np.float32) 
D0 = 40 #cut off frequency

for u in range(rows):
    for v in range(cols):
        D = np.sqrt((u-rows/2)**2 + (v-cols/2)**2)
        if D <= D0:
            H[u,v] = 1
        else:
            H[u,v] = 0
            
plt.imshow(H, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Ideal Low Pass Filtering
Gshift = img_fft * H
plt.imshow(np.log1p(np.abs(Gshift)), cmap='gray')
plt.axis('off')
plt.show()

# Inverse Fourier Transform
G = np.fft.ifftshift(Gshift)
plt.imshow(np.log1p(np.abs(G)), cmap='gray')
plt.axis('off')
plt.show()

g = np.abs(np.fft.ifft2(G))
# plt.imshow(g, cmap='gray')
# plt.axis('off')
# plt.show()
```

```python
cv2_imshow(g)
```

```python
# Filter: High pass filter
H = 1 - H

plt.imshow(H, cmap='gray')
plt.axis('off')
plt.show()

# Ideal High Pass Filtering
Gshift = img_fft * H
plt.imshow(np.log1p(np.abs(Gshift)), 
           cmap='gray')
plt.axis('off')
plt.show()

# Inverse Fourier Transform
G = np.fft.ifftshift(Gshift)
plt.imshow(np.log1p(np.abs(G)), 
           cmap='gray')
plt.axis('off')
plt.show()

g = np.abs(np.fft.ifft2(G))
```

```python
cv2_imshow(g)
```
