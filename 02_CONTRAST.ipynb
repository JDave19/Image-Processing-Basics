```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
from google.colab.patches import cv2_imshow
```

```python
img  = cv2.imread('/content/drive/MyDrive/Colab Notebooks/lenna.png')
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
```

```python
r1 = r*0.298
g1 = g*0.587
b1 = b*0.114

gs = (r1+g1+b1).astype(float)
cv2_imshow(gs)
```

```python
ma = gs.max()
mi = gs.min()
contrast = 255*(gs-mi)/(ma-mi)
cont = contrast.astype(np.uint8)
cv2_imshow(cont)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121), plt.hist(gs.ravel(),256,[0,256]), plt.title("Before contarsting")
plt.subplot(122), plt.hist(cont.ravel(),256,[0,256]), plt.title("After contarsting")
plt.show()
```
