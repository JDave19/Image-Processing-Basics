```python
import matplotlib.image as img
import numpy as np
from google.colab.patches import cv2_imshow
from math import ceil
import matplotlib.pyplot as plt
import cv2
```

```python
# provide the location of image for reading
m = cv2.imread("/content/drive/MyDrive/Colab Notebooks/lenna.png");
m = 0.289*m[:,:,0] + 0.587*m[:,:,1] + 0.114*m[:,:,2]

# determining the length of original image
w1, h1 = m.shape
plt.imshow(m, cmap = 'gray')
```

```python
m = m*255
m = m.astype(np.uint8)
```

```python
print(m.min(),m.max())
print(m.dtype)
```

```python
# height of image after scaling
w2 = int(w1 * 2)
h2= int(h1 * 2)
print(w2,h2)
```

```python
#finding the ratios between old and new image. gives us the point value. will be used to
x_rat = (w1/w2)
y_rat = (h1/h2)
print(x_rat, y_rat)
```

```python
newImage = np.zeros((w2 ,h2)).astype(np.uint8)
# print(newImage.shape) 
for i in range(newImage.shape[0]-1):
   for j in range(newImage.shape[1]-1):
     new_x = ceil((i*x_rat))
     new_y = ceil((j*y_rat))
     newImage[i,j] = m[new_x,new_y]
```

```python
plt.figure(figsize = (20,20))
plt.subplot(121), plt.imshow(m, cmap = 'gray'), plt.title("Original Image"), plt.axis(xmin = 0, xmax = 1024, ymin = 1024, ymax = 0)
plt.subplot(122), plt.imshow(newImage, cmap = 'gray'), plt.title("Scaled Image")
```
