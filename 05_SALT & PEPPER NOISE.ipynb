```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
import random
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/lenna.png')
gs = 0.289*image[:,:,0] + 0.587*image[:,:,1] + 0.114*image[:,:,2]
cv2_imshow(gs)
```

```python
def noise(img: np.array):
   row , col = img.shape
   number_of_pixels = random.randint(100, 10000)
   for i in range(number_of_pixels):
     # Picking random x,y coordinate for adding black and white pixels 0,255
     y_coord=random.randint(0, img.shape[0]-1)
     x_coord=random.randint(0, img.shape[1]-1)
     img[y_coord][x_coord] = 255

   for i in range(number_of_pixels):
     y_coord=random.randint(0, img.shape[0]-1)
     x_coord=random.randint(0, img.shape[1]-1)
     img[y_coord][x_coord] = 0
         
   return img
```

```python
noisy_img = noise(gs)
cv2_imshow(noisy_img)
```
